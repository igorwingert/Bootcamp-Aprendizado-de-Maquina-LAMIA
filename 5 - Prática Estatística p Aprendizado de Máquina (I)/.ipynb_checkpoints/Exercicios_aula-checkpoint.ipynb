Exercicio Probabilidade Condicional

```python
from numpy import random
random.seed(0)

totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0} # cria o dicionario de totais
purchases = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0} # cria o dicionario de compras
totalPurchases = 0 # inicia o total de compras
for _ in range(100000): # percorre 100000 pessoas
    ageDecade = random.choice([20, 30, 40, 50, 60, 70]) # escolhe uma faixa de idade
    purchaseProbability = float(ageDecade) / 100.0 # calcula a probabilidade de compra
    totals[ageDecade] += 1 # soma uma pessoa na faixa
    if (random.random() < purchaseProbability): # ve se a pessoa comprou
        totalPurchases += 1 # soma no total de compras
        purchases[ageDecade] += 1 # soma uma compra na faixa
```

```python
totals # mostra os totais por faixa
```

```python
purchases
```

```python
totalPurchases # total de compras
```

```python
PEF = float(purchases[30]) / float(totals[30]) # calcula a probabilidade de compra dado 30s
print("P(purchase | 30s): " + str(PEF))
```

```python
PF = float(totals[30]) / 100000.0 # calcula a probabilidade de estar nos 30s
print("P(30's): " + str(PF))
```

```python
PE = float(totalPurchases) / 100000.0 # calcula a probabilidade geral de compra
print("P(Purchase):" + str(PE))
```

```python
print("P(30's, Purchase)" + str(float(purchases[30]) / 100000.0)) #mostra a probabilidade conjunta
```

```python
print("P(30's)P(Purchase)" + str(PE * PF)) # mostra o produto das probabilidades
```

```python
print((purchases[30] / 100000.0) / PF) # relacao da probabilidade condicional
```

OJBETIVO: Alterar o código para que a probabilidade de compra não dependa mais da idade, ou seja, em vez de usar uma probabilidade diferente para cada faixa etária, usar uma probabilidade fixa

```python
from numpy import random # importa random do numpy
random.seed(0) # fixa a semente aleatoria

totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
purchases = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0}
totalPurchases = 0 
for _ in range(100000): 
    ageDecade = random.choice([20, 30, 40, 50, 60, 70]) 
    purchaseProbability = 0.4  # fixa a probabilidade fixa de compra
    totals[ageDecade] += 1 
    if (random.random() < purchaseProbability): #
        totalPurchases += 1 
        purchases[ageDecade] += 1 #
```

```python
PEF = float(purchases[30]) / float(totals[30]) # prob. de de compra dado 30s
print("P(purchase | 30s): " + str(PEF))
```

```python
PE = float(totalPurchases) / 100000.0 # prob. geral de compra
print("P(Purchase):" + str(PE))
```
