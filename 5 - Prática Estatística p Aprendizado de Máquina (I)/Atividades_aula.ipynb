```python
from scipy.stats import expon # importa a distribuicao exponencial
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.001) # cria os valores do eixo x
plt.plot(x, expon.pdf(x)) # plota a funcao de densidade exponencial
```

```python
from scipy.stats import binom # binomial
import matplotlib.pyplot as plt

n, p = 10, 0.5 # define o numero de tentativas e a probabilidade
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p)) # plota a funcao de massa binomial
```

```python
from scipy.stats import poisson # poisson
import matplotlib.pyplot as plt

mu = 500 # define a media
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x, mu)) # plota a funcao de massa de poisson
```

covariance and correlation

```python
import numpy as np
import matplotlib.pyplot as plt

def de_mean(x): # cria uma funcao para remover a media
    xmean = np.mean(x) # calcula a media
    return [xi - xmean for xi in x] # retorna os valores sem a media

def covariance(x, y): # cria a funcao de covariancia
    n = len(x) # pega o tamanho da amostra
    return np.dot(de_mean(x), de_mean(y)) / (n-1) # calcula

pageSpeeds = np.random.normal(3.0, 1.0, 1000) # gera velocidades de pagina
purchaseAmount = np.random.normal(50.0, 10.0, 1000) # gera valores de compra

plt.scatter(pageSpeeds, purchaseAmount) # dispersao

covariance (pageSpeeds, purchaseAmount) # covariancia
```

```python
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds # cria valores relacionados a velocidade

plt.scatter(pageSpeeds, purchaseAmount)

covariance (pageSpeeds, purchaseAmount)
```

```python
def correlation(x, y):
    stddevx = x.std() # dp de x
    stddevy = y.std() # dp de y
    return covariance(x,y) / stddevx / stddevy

correlation(pageSpeeds, purchaseAmount) # correlacao
```

```python
import numpy as numpy
numpy.corrcoef(pageSpeeds, purchaseAmount) # calc a matriz de correlacao
```

```python
purchaseAmount = 100 - pageSpeeds * 3 # cria uma relacao linear negativa
plt.scatter(pageSpeeds, purchaseAmount)
correlation (pageSpeeds, purchaseAmount)
```
