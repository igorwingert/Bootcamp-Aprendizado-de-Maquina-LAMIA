```python
import os
import requests
from bs4 import BeautifulSoup as bs
from urllib.parse import urljoin
```

```python
# para a prática optei pelo scraping de um site que fiz algum tempo atrás (não utilizado atualmente) para informações / confirmação do casamento de um amigo meu

url = "https://igdev-git.github.io/casamento-gabrieleandressa/" # url do site escolhido para a pratica

pasta_destino = r"C:\Users\igwin\Videos\Bootcamp-Aprendizado-de-Maquina---LAMIA\8 - Prática Web Scraping com Python p Ciência de Dados (II)"
pasta_imagens = os.path.join(pasta_destino, "imagens_casamento") # pasta onde os arquivos serao salvos

os.makedirs(pasta_imagens, exist_ok=True)

r = requests.get(url) # carrega a pagina

webpage = bs(r.content, "html.parser") # transforma o conteudo em beautifulsoup

titulo = webpage.find("h1").get_text(strip=True) # pega o titulo principal

subtitulos = webpage.find_all(["h2", "h3"]) # pega todos os subtitulos h2 e h3

paragrafos = webpage.find_all("p") # pega todos os paragrafos

links = webpage.find_all("a") # pega todos os links

imagens = webpage.find_all("img") # pega todas as imagens
```

```python
print("TITULO DO SITE:")
print(titulo)
```

```python
print("\nSUBTITULOS ENCONTRADOS:")
for subtitulo in subtitulos:
    print("-", subtitulo.get_text(strip=True))

print("\nTEXTOS ENCONTRADOS:")
for paragrafo in paragrafos:
    texto = paragrafo.get_text(strip=True)

    if texto != "":
        print("-", texto)

print("\nLINKS ENCONTRADOS:")
for link in links:
    texto_link = link.get_text(strip=True)
    href = link.get("href")

    if href:
        link_completo = urljoin(url, href)
        print("-", texto_link, "->", link_completo) # exibe os subtitulos, textos e links encontrados na pagina do site
```

```python
print("\nBAIXANDO IMAGENS:")

for i, imagem in enumerate(imagens):
    src = imagem.get("src")
    alt = imagem.get("alt")

    if src:
        imagem_url = urljoin(url, src)

        img_data = requests.get(imagem_url).content

        nome_arquivo = f"imagem_{i + 1}.jpg"
        caminho_arquivo = os.path.join(pasta_imagens, nome_arquivo)

        with open(caminho_arquivo, "wb") as handler:
            handler.write(img_data)

        print(f"Imagem {i + 1} salva:", caminho_arquivo)
        print("Descricao:", alt)

print("\npratica finalizada!") # baixa as imagens encontradas no site e salva cada uma localmente
```
