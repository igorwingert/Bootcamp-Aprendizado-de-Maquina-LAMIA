```python
import numpy as np
import matplotlib.pylab as plt
import matplotlib # importa algumas bibliotecas para uso posterior
```

```python
# dados para plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)  # calcula os valores do sinal

fig, ax = plt.subplots()  # cria a figura e os eixos
ax.plot(t, s)  # plota o grafico

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='Simples assim amigos')  # define rotulos e titulo
ax.grid()  #exibe a grade
```
