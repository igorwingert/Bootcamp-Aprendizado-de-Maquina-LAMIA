# Prática no jupyter notebook
nesse notebook, criei uma pequena base de dados local, salvei em csv e depois usei pandas e matplotlib para analisar os dados.
realisticamente, a aula foi mais voltada para os 3 ambientes online, então utilizarei como base mais o aprendizado de outras aulas em diferentes IDEs online

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = {
    'mes': ['jan', 'fev', 'mar', 'abr', 'mai'],
    'vendas': [120, 150, 130, 180, 210],
    'gastos': [80, 90, 95, 100, 120]
}

df = pd.DataFrame(dados)
df # dataset simples
```

```python
df.to_csv('vendas_pratica.csv', index=False)  # salva os dados em csv
```

```python
df = pd.read_csv('vendas_pratica.csv')  # le o arquivo csv local
df.head()
```

```python
df['lucro'] = df['vendas'] - df['gastos']  # calcula o lucro
df
```

```python
df.plot(x='mes', y='lucro', kind='bar', title='Lucro por mes')
```
