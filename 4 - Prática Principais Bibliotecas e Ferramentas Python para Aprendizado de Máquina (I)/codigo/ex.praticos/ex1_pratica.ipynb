Nesse exemplo foi usado o numy para armazenar notas em um array e calcular informações simples, como media, maior nota, menor nota e quantidade de notas acima da média

```python
# pratica simples usando numpy no jupyter
import numpy as np #importa numpy como np
```

```python
# notas de alunos
notas = np.array([8.5, 7.0, 9.0, 6.5, 10.0, 7.5])
```

```python
media = np.mean(notas) #aplica media
maior_nota = np.max(notas) #pega o maior
menor_nota = np.min(notas) #pega o menor
```

```python
notas_acima_media = notas[notas > media] # filtra apenas as notas acima da media
```

```python
print(f"notas cadastradas: {notas}")
print(f"media das notas: {media:.2f}")
print(f"maior nota: {maior_nota}")
print(f"menor nota: {menor_nota}")
print(f"notas acima da media: {notas_acima_media}")
print(f"quantidade de notas acima da media: {len(notas_acima_media)}")
```
