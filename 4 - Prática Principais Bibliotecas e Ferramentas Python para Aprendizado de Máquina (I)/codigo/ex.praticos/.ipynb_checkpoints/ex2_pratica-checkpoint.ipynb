cadastro simples de participante bootcamp, inspirado no meu exercicio do card 2

```python
import pandas as pd #usando pandas
```

```python
dados = {
    "nome": ["Igor", "Gabriel", "Josias", "Amanda"],
    "trilha": ["Machine Learning", "Frontend", "Machine Learning", "Frontend"],
    "cards_concluidos": [12, 8, 20, 15]
}
```

```python
df = pd.DataFrame(dados) #cria o df
```

```python
def verificar_progresso(cards):
    # verifica o progresso de acordo com a quantidade de cards
    if cards >= 20:
        return "avancado"
    elif cards >= 10:
        return "intermediario"
    else:
        return "inicio"
```

```python
df["progresso"] = df["cards_concluidos"].apply(verificar_progresso) #pega cards_concluidos, aplica a func e armazena em progresso
```

```python
print("tabela de participantes:")
display(df) #exibe a tabela
```

```python
print(df["cards_concluidos"].mean()) #media de cards concluidos
```

```python
print(df["trilha"].value_counts()) #qtd de participantes por trilha, utilizando count
```
