***Exercícios resolvidos numpy***

```python
import numpy as np #importa numpy com abreviatura np
```

```python
np.zeros(10) #matriz de 10 zeros
```

```python
np.ones(10) #matriz de 10 uns
```

```python
np.ones(10) * 5 #matriz de 10 cincos (nao exxiste np.fives)
```

```python
np.arange(10, 51) #matriz 10 a 50
```

```python
np.arange(10, 51, 2) #array de numeros pares de 10 até 50
```

```python
np.arange(0,9).reshape (3,3)##cria matriz 3x3 com valores de 0 a 8
```

```python
np.eye(3)#matriz identity 3x3
```

```python
np.random.rand (1) #numeros aleatorios entre 0 e 1
```

```python
np.random.randn(25)#gera um array de 25 num aleatorios de uma distribuição normal
```

```python
np.arange(0, 101) / 100 #criado uma matriz associado a uma op. escalar
```

```python
np.linspace(0,1,20)# array de tam.20 igualmente espaçado entre 0 e 1
```

```python
mat = np.arange(1,26).reshape(5,5)
mat
```

replicação de matrizes conforme está na questão:

```python
mat[2:, 1:]
```

```python
mat[3, -1]#elem 20 da matriz
```

```python
mat[:3,1]#puxa apenas a segunda coluna com os 3 elementos apresentados
```

```python
mat [-1, :]#ultima linha da matriz como vetor
```

```python
mat[3:, :]
```

```python
mat.sum()#soma de todos os valores no mat
```

```python
np.std(mat) #desvio padrao dos valores em mat
```

```python
mat.sum(axis=0) #obtem soma de todas as colunas em mat
```
