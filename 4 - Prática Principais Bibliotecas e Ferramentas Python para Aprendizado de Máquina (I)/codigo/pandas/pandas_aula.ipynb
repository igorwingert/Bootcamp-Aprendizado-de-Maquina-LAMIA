```python
esquerda = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
```

```python
direita = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
esquerda
```

```python
direita
```

```python
pd.merge(esquerda,direita,how='inner',on='key') #mescla
```

```python
esquerda = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
```

```python
direita = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(esquerda, direita, on=['key1', 'key2'])#mescla usando duas chaves
```

```python
pd.merge(esquerda, direita, how='outer', on=['key1', 'key2']) #mantem os registros das duas tabelas
```

```python
pd.merge(esquerda, direita, how='right', on=['key1', 'key2']) #mantem os registros da tabela direita
```

```python
pd.merge(esquerda, direita, how='left', on=['key1', 'key2']) #mantem os da esquerda
```
