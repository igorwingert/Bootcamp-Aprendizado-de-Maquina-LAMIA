```python
def chefe_cargo(cargo):
    if 'chief' in cargo.lower():
        return True
    else:
        return False#pessoas que tem a palavra chefe em seu cargo 
```

```python
sum(sal['JobTitle'].apply(lambda x: chefe_cargo(x))) # lambda cria uma função anônima rápida para aplicar uma operação em cada elemento
```
